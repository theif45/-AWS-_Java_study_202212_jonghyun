```python
import numpy as np

a1 = np.arange(30).reshape(3,5,2)
print(a1, a1.shape)
```

```python
a2 = np.sum(a1, axis=0)
print(a2, a2.shape)
```

```python
a3 = np.sum(a1, axis=2)
print(a3, a3.shape)
```

```python
a4 = np.sum(a1, axis=1)
print(a4,a4.shape)
```

```python
from matplotlib import image

image = image.imread('cat.jpg')
print(image.shape)
```

```python
import matplotlib.pyplot as plt

_, axe = plt.subplots()
axe.imshow(image)
```

```python
image2 = np.moveaxis(image, [2], [0])
print(image2.shape)
```

```python
_, axe = plt.subplots()
axe.imshow(image2[0])
```
