```python
def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list
```

```python
import numpy as np

csv_data=[]
with open('전국로컬푸드인증정보표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line))
print(csv_data)
```
