```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_data = pd.read_csv('내국인출국교통수단별_20230302174035.csv', encoding='cp949', header=1)
pd_data.columns = ['시점','공항','항구']
pd_data
print(pd_data.shape)
```

```python
sub_data2022 = pd_data.tail(12)
print(sub_data2022)
print(sub_data2022.describe())

for n in ['공항','항구']:
    value_min = sub_data2022[n].min()
    value_max = sub_data2022[n].max()
    value_std = sub_data2022[n].std()
    value_mean = sub_data2022[n].mean()
    sub_data2022[n + '_norm'] = sub_data2022[n].apply(lambda x: (x-value_min)/(value_max-value_min))
    sub_data2022[n + '_std'] = sub_data2022[n].apply(lambda x: (x-value_mean)/value_std)

_, axe = plt.subplots()
axe.plot(sub_data2022['공항'], color='blue')
axe.plot(sub_data2022['항구'], color='red')

_, axe = plt.subplots()
axe.plot(sub_data2022['공항_norm'], color='blue')
axe.plot(sub_data2022['항구_norm'], color='red')

_, axe = plt.subplots()
axe.plot(sub_data2022['공항_std'], color='blue')
axe.plot(sub_data2022['항구_std'], color='red')
```

```python
print(pd_data)
for i in range(8):
    #print(i, 12*i, 12*(i+1))
    sub_data = pd_data.iloc[12*i:12*(i+1)]
    print(sub_data)
    
    for n in ['공항','항구']:
        value_min = sub_data[n].min()
        value_max = sub_data[n].max()
        value_std = sub_data[n].std()
        value_mean = sub_data[n].mean()
        sub_data[n + '_norm'] = sub_data[n].apply(lambda x: (x-value_min)/(value_max-value_min))
        sub_data[n + '_std'] = sub_data[n].apply(lambda x: (x-value_mean)/value_std)

    fig, axe = plt.subplots(1,3)
    fig.set_figwidth(6.4*3)
    
    axe[0].plot(np.arange(1,13),sub_data['공항'], color='blue')
    axe[0].plot(np.arange(1,13),sub_data['항구'], color='red')
    axe[0].set_title('{}_raw'.format(2015+i))

    axe[1].plot(np.arange(1,13),sub_data['공항_norm'], color='blue')
    axe[1].plot(np.arange(1,13),sub_data['항구_norm'], color='red')
    axe[1].set_title('{}_norm'.format(2015+i))

    axe[2].plot(np.arange(1,13),sub_data['공항_std'], color='blue')
    axe[2].plot(np.arange(1,13),sub_data['항구_std'], color='red')
    axe[2].set_title('{}_std'.format(2015+i))
    
```
