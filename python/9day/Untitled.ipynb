```python
import numpy as np

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

csv_data = []
with open('전국로컬푸드인증정보표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line))

location_list = []
item_list = []
for i, data in enumerate(csv_data[1:]):
    location = data[4]
    items = data[8].replace('"','').replace(',','+').replace(' ','').split('+')
    items = list(set(items))
    for item in items:
        item_list.append(item)
        location_list.append(location)
        
item_to_location_dict = dict(zip(item_list, location_list))
unique_item_list = set(item_list)

location_to_item_dict = {}
for item in unique_item_list:
    location_to_item_dict[item] = []
    for key, value in item_to_location_dict.items():
        if item == key:
            location_to_item_dict[item].append(value)

item_frequency = {}
for item in unique_item_list:
    item_frequency[item] = item_list.count(item)
item_frequency = dict(sorted(item_frequency.items(), key=lambda x:x[1], reverse=True))

top_10_items = {}
i = 0
for key, value in item_frequency.items():
    if i >= 10:
        break
    top_10_items[key] = value
    i += 1

print("Top 10 popular items (count by item)")
print(top_10_items)
print("\nDictionary created: {County unit area name: [item name]}")
print(item_to_location_dict)
print("\nDictionary created: {Item: {[County unit area name]}}")
print(location_to_item_dict)
```
