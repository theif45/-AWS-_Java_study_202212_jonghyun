```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data'
pd_data = pd.read_csv(url, header=None)
pd_data.columns = ['구매가', '유지비용','문개수','탑승인원','수납공간','안전도','만족도']
pd_data.head()
```

* 구매가가 만족도에 미치는 영향
    - figure를 좌우로 나누어, 왼쪽은 구매가 vs.만족도 산점도, 오른쪽은 아래위로 분할, 위쪽은 구매가 vhigh 중 만족도 count, 아래쪽은 구매가 low중 만족도별 count
* 수납공간이 크면 탑승인원도 많은가

```python
fig = plt.figure(figsize=(12,8))
a = fig.add_subplot(1,2,1)
b = fig.add_subplot(2,2,2)
c = fig.add_subplot(2,2,4)

a.scatter(pd_data['구매가'],pd_data['만족도'])

filter1 = pd_data['구매가'] == 'vhigh'
b.bar(pd_data.loc[filter1,'만족도'].unique(),pd_data.loc[filter1,'만족도'].value_counts())

filter2 = pd_data['구매가'] == 'low'
c.bar(pd_data.loc[filter2,'만족도'].unique(),pd_data.loc[filter2,'만족도'].value_counts())
```

```python
fig = plt.figure(figsize=(5,5))
d = fig.add_subplot(1,1,1)

filter3 = pd_data['수납공간'] == 'big'
d.bar(pd_data.loc[filter3,'탑승인원'].unique(),pd_data.loc[filter3,'탑승인원'].value_counts())
```
