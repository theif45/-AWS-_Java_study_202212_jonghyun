```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_tips = sns.load_dataset('tips')
print(df_tips)
```

```python
# hue, style, size
# sns.relplot(data=df_tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
# row, col
sns.relplot(data=df_tips, x='total_bill', y='tip', col='sex', row='smoker', hue='size', size='size')
```

```python
x = sns.relplot(data=df_tips, x='total_bill', y='tip', kind='line')
print(x.figure, type(x.figure))
x.figure.set_facecolor('skyblue')

print(x.axes, type(x.axes[0][0]))
```

```python
fig1 = plt.figure()
a = fig1.add_subplot()
# sns.scatterplot(data=df_tips,x='total_bill', y='tip', ax=a, hue='day')
sns.lineplot(data=df_tips,x='total_bill', y='tip', ax=a, hue='day')
```
