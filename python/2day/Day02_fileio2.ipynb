```python
data_list = []
with open('전국건강증진센터표준데이터.csv') as f:
    for line in f:
        data_list.append(line.split(','))
data_list = data_list[1:]
print(data_list)
```

```python
do_list = []
adress = []
for i in data_list:
    adress.append(i[2])
print(adress)
print(adress.index(''))
del adress[132]

for i in adress:
    ads = i
    ads_list = ads.split()
    do_list.append(ads_list[1])
print(do_list)
```

```python
do_set = set(do_list)
d_do_list = []
d_do_count = []
for i in do_set:
    d_do_list.append(i)
    d_do_count.append(do_list.count(i))
d1 = dict(zip(d_do_list, d_do_count))
print(d1)
```

```python
d1_sorted = dict(sorted(d1.items(), key = lambda x : x[1]))
print(d1_sorted)
```
