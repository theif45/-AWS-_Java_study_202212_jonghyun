```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_hydrant = pd.read_csv('전국소방용수시설표준데이터.csv',encoding='cp949')
pd_parking = pd.read_csv('전국소방자동차전용구역표준데이터.csv',encoding='cp949')
```

```python
pd_hydrant = pd_hydrant[['시도명','소재지도로명주소','위도','경도']]
pd_parking = pd_parking[['시도명','소재지도로명주소','위도','경도']]
```

```python
df_fire = pd.concat((pd_hydrant,pd_parking), keys=['hydrant','parking'])
```

```python
df_fire.reset_index(inplace=True)
df_fire.drop('level_1',axis=1,inplace=True)
cname_list = list(df_fire.columns)
cname_list[0] = '종류'
df_fire.columns = cname_list
df_fire
```

```python
g1 = df_fire.groupby(['시도명','종류'])
g1.count()

g2 = df_fire.groupby(['종류','시도명'])
g2.count()
```
