```python
pd_data = pd.read_csv('이륜차신고현황_시도별_20230309170634.csv', encoding='cp949')
col_name = ['시도명','시점','시군구','경형(계)','소형(계)','중형(계)','대형(계)','경형(관용)','소형(관용)','중형(관용)','대형(관용)','경형(자가용)','소형(자가용)','중형(자가용)','대형(자가용)']
pd_data = pd_data.iloc[1:,:].reset_index(drop=True)
pd_data.columns = col_name
pd_data
```

```python
fig = plt.figure(figsize=(6.4*3,4.8*3),layout='constrained')
a = fig.add_subplot(2,2,1)
b = fig.add_subplot(2,2,2)
c = fig.add_subplot(2,2,3)
d = fig.add_subplot(2,2,4)

sns.barplot(x = pd_data['시군구'],y = pd_data['경형(계)'].astype(np.float64), ax=a)
sns.barplot(x = pd_data['시군구'],y = pd_data['소형(계)'].astype(np.float64), ax=b)
sns.barplot(x = pd_data['시군구'],y = pd_data['중형(계)'].astype(np.float64), ax=c)
sns.barplot(x = pd_data['시군구'],y = pd_data['대형(계)'].astype(np.float64), ax=d)
a.set_xticklabels(labels=pd_data['시군구'], fontsize=10, rotation=45)
b.set_xticklabels(labels=pd_data['시군구'], fontsize=10, rotation=45)
c.set_xticklabels(labels=pd_data['시군구'], fontsize=10, rotation=45)
d.set_xticklabels(labels=pd_data['시군구'], fontsize=10, rotation=45)
a.set_yticks(np.linspace(0,12000,7))
b.set_yticks(np.linspace(0,12000,7))
c.set_yticks(np.linspace(0,12000,7))
d.set_yticks(np.linspace(0,12000,7))

e = fig.add_axes([0.2,0.2,0.02,0.02])
f = fig.add_subplot(6,6,4)
g = fig.add_subplot(6,6,19)
h = fig.add_subplot(6,6,22)

sns.barplot(x = pd_data['시군구'],y = pd_data['경형(관용)'].astype(np.float64), ax=e)
sns.barplot(x = pd_data['시군구'],y = pd_data['소형(관용)'].astype(np.float64), ax=f)
sns.barplot(x = pd_data['시군구'],y = pd_data['중형(관용)'].astype(np.float64), ax=g)
sns.barplot(x = pd_data['시군구'],y = pd_data['대형(관용)'].astype(np.float64), ax=h)
e.set_xticklabels(labels='', fontsize=10, rotation=45)
f.set_xticklabels(labels='', fontsize=10, rotation=45)
g.set_xticklabels(labels='', fontsize=10, rotation=45)
h.set_xticklabels(labels='', fontsize=10, rotation=45)
e.set_yticklabels(labels='')
f.set_yticklabels(labels='')
g.set_yticklabels(labels='')
h.set_yticklabels(labels='')
```
