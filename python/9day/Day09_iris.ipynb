```python
import requests
import numpy as np

x = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
```

```python
raw_data=[]
for line in x.text.split('\n'):
    raw_data.append(line.split(','))
    
for sample in raw_data:
    if len(sample)<1:
        continue
    for i, value in enumerate(sample[:4]):
        try:
            sample[i] = float(value)
        except:
            sample[i] = 0.0
```

```python
for sample in raw_data:
    if len(sample)<2:
        continue
    if 'setosa' in sample[-1]:
        sample[-1] = 0.0
    elif 'versicolor' in sample[-1]:
        sample[-1] = 1.0
    elif 'virginica' in sample[-1]:
        sample[-1] = 2.0
```

```python
raw_data = raw_data[:150]
print(raw_data)
print(len(raw_data))
```

```python
np_data = np.array(raw_data)
np_data.shape
np_data[:3]
```

### 기본 통계값
* 전체(150샘플) 평균, 표준편차 - sepal length, sepal width, petal length, petal width
* 품종별 평균, 표준편차
* 평균 np.mean()
* 표준편차 np.std()

```python
total_mean = np.mean(np_data[:,:4],axis=0)
total_std = np.std(np_data[:,:4],axis=0)
print(total_mean)
print(total_std)
print(np.corrcoef(np_data.T))
```

```python
filter_setosa = np_data[np_data[:,-1] == 0.0]
filter_versicolor = np_data[np_data[:,-1] == 1.0]
filter_virginica = np_data[np_data[:,-1] == 2.0]
```

```python
setosa_mean = np.mean(filter_setosa,axis=0)
versicolor_mean = np.mean(filter_versicolor,axis=0)
virginica_mean = np.mean(filter_virginica,axis=0)

setosa_std = np.std(filter_setosa,axis=0)
versicolor_std = np.std(filter_versicolor,axis=0)
virginica_std = np.std(filter_virginica,axis=0)
print('setosa',setosa_mean,setosa_std)
print('versicolor',versicolor_mean,versicolor_std)
print('virginica',virginica_mean,virginica_std)
```
