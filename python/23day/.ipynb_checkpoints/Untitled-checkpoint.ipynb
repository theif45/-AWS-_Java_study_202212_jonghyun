```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

class GraphGUI:
    def __init__(self, master):
        self.master = master
        master.title("Graph GUI")
        master.geometry("400x400")
        # 파일 선택 버튼
        self.file_button = tk.Button(master, text="Select File", command=self.load_file)
        self.file_button.pack(padx=10, pady=10)

        # x축, y축 데이터 선택 버튼
        self.x_label = tk.Label(master, text="Select X Axis Label:")
        self.x_label.pack()
        self.x_options = tk.StringVar(master)
        self.x_options.set("")
        self.x_dropdown = tk.OptionMenu(master, self.x_options, "")
        self.x_dropdown.pack()
        self.y_label = tk.Label(master, text="Select Y Axis Label:")
        self.y_label.pack()
        self.y_options = tk.StringVar(master)
        self.y_options.set("")
        self.y_dropdown = tk.OptionMenu(master, self.y_options, "")
        self.y_dropdown.pack()

        # 그래프 출력 버튼
        self.plot_button = tk.Button(master, text="Plot Graph", command=self.plot_graph)
        self.plot_button.pack(padx=10, pady=10)

    # 파일 불러오기 함수
    def load_file(self):
        file_path = filedialog.ask(fiopenfilenameletypes=[("CSV files", "*.csv")])
        if file_path:
            try:
                # 파일 불러오기
                data = pd.read_csv(file_path, encoding='cp949', header=1)
                # x축, y축 데이터 선택 목록 갱신
                self.x_options.set("")
                self.y_options.set("")
                self.x_options.set(list(data.columns))
                self.y_options.set(list(data.columns))
            except Exception as e:
                tk.messagebox.showerror("Error", str(e))

    # 그래프 출력 함수
    def plot_graph(self):
        try:
            # 선택한 x축, y축 데이터로 그래프 작성
            x_label = self.x_options.get()
            y_label = self.y_options.get()
            data = pd.read_csv(file_path, encoding='cp949', header=1)
            fig, ax = plt.subplots()
            ax.plot(data[x_label], data[y_label], label='data', color='blue', linestyle='-', linewidth=3)
            ax.legend()
            ax.set_title('Graph Title')
            ax.set_xlabel(x_label)
            ax.set_ylabel(y_label)
            # 그래프 출력
            canvas = FigureCanvasTkAgg(fig, master=root)
            canvas.get_tk_widget().pack()
        except Exception as e:
            tk.messagebox.showerror("Error", str(e))

root = tk.Tk()
my_gui = GraphGUI(root)
root.mainloop()
```
