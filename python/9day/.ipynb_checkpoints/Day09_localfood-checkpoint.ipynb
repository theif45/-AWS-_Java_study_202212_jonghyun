### 전국로컬푸드인증정보표준데이터

url: https://www.data.go.kr/data/15025455/standard.do


* 인기품목 top 10(품목별 카운트)
** 다중품목(+또는,로 연결)은 각각 분리
** 이외 단일품목으로 취급
** 나머지는 알아서
* dictionary 작성: {군단위 지역명: [품목명]}
* dictionary 작성: {품목 : {[군단위 지역명]}

* jypyter note 업로드: 이름_로컬푸드.jpynp

```python
def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list
```

```python
import numpy as np

csv_data=[]
with open('전국로컬푸드인증정보표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line))
print(csv_data[:3])
```

```python
length=[]
for e in csv_data:
    length.append(len(e))
print(set(length))
```

```python
for i,e in enumerate(csv_data[0]):
    print(i,e)
# 4 사업장도로명주소
# 8 품목명
```
