```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_oneway = pd.read_csv('전국일방통행도로표준데이터_pass1.csv')
pd_child = pd.read_csv('전국어린이보호구역표준데이터_pass1.csv')
pd_elder = pd.read_csv('전국노인장애인보호구역표준데이터_pass1.csv')
```

```python
print(pd_oneway.head())    # 시도명 별 보차분리여부 y count
print(pd_child.head())    # 시도명(지자체명) 별 CCTV 설치대수 count
print(pd_elder.head())    # 시도명 별 CCTV 설치대수 count
```

```python
g1 = pd_oneway[pd_oneway['보차분리여부']=='Y'].groupby('시도명')
s1 = g1['보차분리여부'].count()
s1.name = '일방통행보차분리Y_COUNT'
print(s1)
```

```python
g2 = pd_child.groupby('지자체명')
s2 = g2['CCTV설치대수'].sum()
s2.name = '어린이CCTV_COUNT'
print(s2)
```

```python
g3 = pd_elder.groupby('시도명')
s3 = g3['CCTV설치대수'].sum()
s3.name = '노인장애인CCTV_COUNT'
print(s3)
```

```python
df_combined = pd.concat((s1,s2,s3), axis=1)
print(df_combined)
```

```python
df_combined.reset_index(inplace=True)
```

```python
 _,axe = plt.subplots()
axe.plot(df_combined['어린이CCTV_COUNT'])
axe.plot(df_combined['노인장애인CCTV_COUNT'])
```

```python
# normalization : (x-x_min)/(x_max-x_min)
df_combined['어린이CCTV_COUNT'].describe()
chlid_min = df_combined['어린이CCTV_COUNT'].min()
child_max = df_combined['어린이CCTV_COUNT'].max()
df_combined['어린이CCTV_COUNT_norm'] = df_combined['어린이CCTV_COUNT'].apply(lambda x: (x-chlid_min)/(child_max-chlid_min))

df_combined['노인장애인CCTV_COUNT'].describe()
elder_min = df_combined['노인장애인CCTV_COUNT'].min()
elder_max = df_combined['노인장애인CCTV_COUNT'].max()
df_combined['노인장애인CCTV_COUNT_norm'] = df_combined['노인장애인CCTV_COUNT'].apply(lambda x: (x-elder_min)/(elder_max-elder_min))

print(df_combined['노인장애인CCTV_COUNT_norm'].describe())

_, axe = plt.subplots()
axe.plot(df_combined['어린이CCTV_COUNT_norm'], color='blue')
axe.plot(df_combined['노인장애인CCTV_COUNT_norm'], color='red')
```

```python
# standardization : (x - x_mean) / x_std
df_combined['어린이CCTV_COUNT'].describe()
child_mean = df_combined['어린이CCTV_COUNT'].mean()
child_std = df_combined['어린이CCTV_COUNT'].std()
df_combined['어린이CCTV_COUNT_std'] = df_combined['어린이CCTV_COUNT'].apply(lambda x: (x-child_mean)/(child_std))

elder_mean = df_combined['노인장애인CCTV_COUNT'].mean()
elder_std = df_combined['노인장애인CCTV_COUNT'].std()
df_combined['노인장애인CCTV_COUNT_std'] = df_combined['노인장애인CCTV_COUNT'].apply(lambda x: (x-elder_mean)/(elder_std))

print(df_combined['노인장애인CCTV_COUNT_std'].describe())

_, axe = plt.subplots()
axe.plot(df_combined['어린이CCTV_COUNT_std'], color='blue')
axe.plot(df_combined['노인장애인CCTV_COUNT_std'], color='red')
```
