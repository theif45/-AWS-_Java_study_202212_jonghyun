```python
import numpy as np
import matplotlib.pyplot as plt
import math

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

def split_len(data_list):
    len_list=[]
    for e in data_list:
        len_list.append(len(e))
    print(set(len_list))
    if len(set(len_list))>1:
        for i in set(len_list):
            print(i, len_list.count(i))
```

```python
csv_data=[]
with open('가스공급량_20230217170919.csv') as f:
    for line in f:
        csv_data.append(my_split(line[:-1]))
np_data = np.array(csv_data)
print(np_data)
for i in enumerate(np_data[1]):
    print(i)
```

```python
np_data = np_data[:,[0,1,3,5,7]]
np_data = np_data[np_data[:,0]!='"계"']
print(np_data)
```

```python
np_data = np_data[1:]
np_data[0][0] = '월'
np_data[0][1] = '년도'
np_data[0][2] = '도시가스 판매량'
np_data[0][3] = '프로판 판매량'
np_data[0][4] = '부탄 판매량'
for i,e in enumerate(np_data[:,0]):
    np_data[i,0] = e.replace('"','')
for i,e in enumerate(np_data[:,1]):
    np_data[i,1] = e.replace('"','')
print(np_data[:3])
```
