```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
xs = np.linspace(0, 50, 100)
ys = np.sin(xs)

# print(xs)
# print(ys)

figure, axe = plt.subplots()
figure.set_facecolor('skyblue')
figure.suptitle('Fiqure1')
figure.supxlabel('Figuer1_x')
figure.supylabel('Figuer1_y')
figure.text(0.5,0.7,'test')
figure.set_size_inches(15,4.8)
axe.plot(xs,ys)
axe.set_title('my axe title')
axe.set_xlabel('axe_x')
axe.set_ylabel('axe_y')
axe.set_xlim(0,30)
axe.set_ylim(0,1)
axe.set_xticks(np.linspace(0,30,20))
# axe.set_xticklabels(list('abcde'))
print(type(figure))
print(type(axe))
```

```python
figure, axe = plt.subplots(2,3, sharey=True, figsize=(15,7))
axe[1,0].plot(xs,ys)
```
