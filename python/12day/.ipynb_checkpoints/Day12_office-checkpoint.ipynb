```python
import numpy as np
import math

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

def split_len(data_list):
    len_list=[]
    for e in data_list:
        len_list.append(len(e))
    print(set(len_list))
    if len(set(len_list))>1:
        for i in set(len_list):
            print(i, len_list.count(i))

def dist_np(p1, p2): #  [3,10]  [5,25]
    return math.sqrt(sum((p2-p1)**2))
```

* oo공인중개... 중 흔한 상호명

```python
csv_data=[]
with open('전국공인중개사사무소표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line[:-1]))
print(csv_data[:5])
split_len(csv_data)
```

```python
for i,e in enumerate(csv_data[0]):
    print(i,e)
```

```python
pass1 = []
pass2 = []
pass3 = []
for e in csv_data[1:]:
    pass1.append(e[2])
    pass2.append(e[7])
    pass3.append(e[0])
```

```python
count1 = pass1.count('공인중개사')
count2 = pass1.count('개업공인중개사')
count3 = pass1.count('법인')
        
print('공인중개사',count1,'객업공인중개사',count2,'법인',count3)
print('공인중개사 비율', count1*100/(count1+count2+count3))
print('객업공인중개사 비율', count2*100/(count1+count2+count3))
```

```python
countNo = pass2.count('N')
countYes = pass2.count('Y')

print('가입',countYes,'미가입',countNo)
print('가입 비율', countYes*100/(countYes+countNo))
print('가입 비율', countNo*100/(countYes+countNo))
```

```python
pass4 = []
pass5 = []
for e in pass3:
    if '공인중개사' in e:
        a = e.replace('공인중개사사무소','').replace('공인중개사사무','').replace('공인중개사','')
        pass4.append(a)
    else:
        pass5.append(e)
not_list = list(set(pass5))
print(len(not_list))
print(not_list[:3])
```

```python
pass4_set = list(set(pass4))
count_name_list=[]
for e in pass4_set:
    count_name_list.append(pass4.count(e))
p_dict = dict(sorted(zip(pass4_set,count_name_list), key=lambda x:x[1], reverse=True))
p_list = list(p_dict.items())
print(len(p_list))
p_list
```
