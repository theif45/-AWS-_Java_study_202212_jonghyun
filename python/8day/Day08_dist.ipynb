```python
import numpy as np
import math

x1=3
y1=10

x2=5
y2=25

def dist(p1,p2):
    return math.sqrt((p2[0]-p1[0])**2 + (p2[1]-p1[1])**2)

b = dist([x1,y1], [x2,y2])
print(b)

def dist_np(p1,p2):
    return math.sqrt(sum((p2-p1)**2))
    
c = dist_np(np.array([x1,y1]),np.array([x2,y2]))
print(c)
```

```python
data=[]
with open('전국자동차정비업체표준데이터.csv') as f:
    for line in f:
        data.append(line.split('\n'))
```

```python
pass1=[]
for e in data:
    a=e[0].replace(',,',',0,').replace('"','0')
    pass1.append(a.split(','))
print(pass1)
```

```python
pass2=[]
pass3=[]
for e in pass1:
    pass2.append(e[4:6])
pass3 = pass2[1:]
print(pass3)
```

```python
for e in pass3:
    for i,s in enumerate(e):
        try:
            e[i]=float(s)
        except:
            if(len(e[i])<3):
               continue
            if(type(e[i]) == str):
                continue
print(pass3)
```

```python
current_x = input()
current_y = input()
for i,e in pass3:
    distance = dist_np(np.array([current_x,current_y],np.float64),np.array(e))
    print(distance)
```
