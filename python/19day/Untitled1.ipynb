```python
import turtle
import random

# 화면 설정
screen = turtle.Screen()
screen.bgcolor('black')
screen.title('간단한 게임')

# 경계선 그리기
border_pen = turtle.Turtle()
border_pen.speed(0)
border_pen.color('white')
border_pen.penup()
border_pen.setposition(-300, -300)
border_pen.pendown()
border_pen.pensize(3)
for side in range(4):
    border_pen.fd(600)
    border_pen.lt(90)
border_pen.hideturtle()

# 플레이어 설정
player = turtle.Turtle()
player.color('blue')
player.shape('triangle')
player.penup()
player.speed(0)
player.setposition(0, -250)
player.setheading(90)

# 플레이어 이동
player_speed = 15
def move_left():
    x = player.xcor()
    x -= player_speed
    if x < -280:
        x = -280
    player.setx(x)

def move_right():
    x = player.xcor()
    x += player_speed
    if x > 280:
        x = 280
    player.setx(x)

turtle.listen()
turtle.onkey(move_left, 'Left')
turtle.onkey(move_right, 'Right')

# 적 설정
enemy = turtle.Turtle()
enemy.color('red')
enemy.shape('circle')
enemy.penup()
enemy.speed(0)
enemy.setposition(random.randint(-250, 250), random.randint(200, 250))

enemy_speed = 2

# 충돌 감지
def is_collision(t1, t2):
    distance = t1.distance(t2)
    if distance < 15:
        return True
    else:
        return False

# 게임 루프
while True:
    # 적 이동
    x = enemy.xcor()
    y = enemy.ycor()
    y -= enemy_speed
    enemy.sety(y)
    
    # 적 위치 초기화
    if y < -250:
        x = random.randint(-250, 250)
        y = random.randint(200, 250)
        enemy.setposition(x, y)

    # 충돌 체크
    if is_collision(player, enemy):
        player.hideturtle()
        enemy.hideturtle()
        print('Game Over')
        break

turtle.done()
```
