```python
import numpy as np
import math
def dist_np(p1,p2):
    return math.sqrt(sum((p2-p1)**2))
def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list
```

```python
csv_data=[]
with open('전국자동차정비업체표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line))                        
print(csv_data[:3])
a=[]
for i in csv_data:
    a.append(len(i))
set(a)
```
