```python
list_a = [i for i in range(10)]
list_b = [i for i in range(100,110)]
print(list_a)
print(list_b)

# in
# in 앞의 값이 있는지 확인
if 10 not in list_a:
    print('True')
else:
    print('False')
    
# concat
# 이어붙일때 사용
list_c = list_a + list_b
print(list_c)

# *
# *한 수만큼 반복해서 리스트에 작성
print(list_a * 2)
```

```python
list_d = list('abcdefghicjklmn')
print(list_d)
print(list_d[3])
print(list_d[3:10:2])
print(len(list_d))
print(min(list_d))
print(max(list_d))
c_index = list_d.index('c',3,10)
print(c_index)
#list_d[c_index]='x'
#print(list_d)
print('count:', list_d.count('c'))
```

```python
s = 'While The Python Language Reference describes the exact syntax and semantics of the Python language, this library reference manual describes the standard library that is distributed with Python. It also describes some of the optional components that are commonly included in Python distributions.'
print(s.lower())
list_s = list(s.lower())
set_s = set(list_s)
#print(list_s)
print(list_s.count('a'))
print(list_s.index('a'))

count = 0
for i in list_s:
    count += 1
    if i == 'a':
        print(count - 1)

print(set_s)
exclude_list = ['.', ' ',',']
for i in set_s:
    if i not in exclude_list:
        print(i, list_s.count(i))
```

```python
list_a = [1,2,3]
list_b = list('abc')
print(list_a)
print(list_b)

# dict 형태로 변환
d2 = dict(zip(list_b, list_a))
print(d2)
print(d2.items())

d3 = dict([('a',1), ('b',2), ('c', 3)])
print(d3)

print(d3['c'])
print(d3.get('c'))
print(d3.get('x'))
d3['y'] = 15
print(d3)
```

```python
for i in d3:
    print(i, d3[i])
    
print(d3.keys())
print(d3.values())

for i in d3.values():
    print(i)
    
print(d3)
d3['y']=20
print(d3)
```
