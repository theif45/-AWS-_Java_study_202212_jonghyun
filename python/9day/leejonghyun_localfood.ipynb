### 전국로컬푸드인증정보표준데이터

url: https://www.data.go.kr/data/15025455/standard.do


* 인기품목 top 10(품목별 카운트)
** 다중품목(+또는,로 연결)은 각각 분리
** 이외 단일품목으로 취급
** 나머지는 알아서
* dictionary 작성: {군단위 지역명: [품목명]}
* dictionary 작성: {품목 : {[군단위 지역명]}

* jypyter note 업로드: 이름_로컬푸드.jpynp

```python
def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list
```

```python
import numpy as np

csv_data=[]
with open('전국로컬푸드인증정보표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line))
print(csv_data[:3])
```

```python
length=[]
for e in csv_data:
    length.append(len(e))
print(set(length))
```

```python
for i,e in enumerate(csv_data[0]):
    print(i,e)
# 4 사업장도로명주소
# 8 품목명
```

```python
pass1 = []
ad=[]
for s in csv_data[1:]:
#     print(s[8])
    ad.append(s[4])
    pass1.append(s[8].replace('"','').replace(',','+').replace(' ','').split('+'))
print(ad)
print(pass1)
```

```python
dict1 = dict(zip(ad,pass1))
print(dict1)
```

```python
pass2=[]
ad2=[]
for i,s in enumerate(pass1):
    for e in s:
        ad2.append(ad[i])
        pass2.append(e)
# print(pass2)
# print(ad2)
dict2 = dict(zip(pass2,ad2))
print(dict2)
```

```python
pass3=set(pass2)
print(pass3)
```

```python
k_data=[]
v_data=[]
for i,e in enumerate(pass3):
    k_data.append(e)
    v_data.append(pass2.count(e))
dict_data = dict(zip(k_data,v_data))
dict_data = dict(sorted(dict_data.items(), key=lambda x:x[1], reverse=True))
print(dict_data)
```
