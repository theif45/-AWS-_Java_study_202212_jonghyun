```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_hydrant = pd.read_csv('전국소방용수시설표준데이터.csv',encoding='cp949')
pd_parking = pd.read_csv('전국소방자동차전용구역표준데이터.csv',encoding='cp949')

print(pd_hydrant.head())
print(pd_parking.head())
```

```python
pd_hydrant = pd_hydrant[['시도명','소재지도로명주소','위도','경도']]
pd_parking = pd_parking[['시도명','소재지도로명주소','위도','경도']]
```

```python
df_fire = pd.concat((pd_hydrant,pd_parking), axis=0, keys=['hydrant','parking'])
df_fire.reset_index(level=0, inplace=True, names='종류')
df_fire
```

```python
df_fire.reset_index(level=0, inplace=True)
df_fire = df_fire[['종류','시도명','소재지도로명주소','위도','경도']]
df_fire
```

```python
g1 = df_fire.groupby('시도명')
s1 = g1['종류'].value_counts()
s1
```
