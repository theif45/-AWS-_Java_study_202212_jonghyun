## 자동차 연비 데이터 정제

uci repo
url: https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data

* requests 패키지 이용 데이터 로딩.
* 최대한 많이 정제

```python
import requests
x = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data')
print(x.text)
```

```python
data=[]
l_data=[]
for line in x.text.split('\n'):
    data.append(line.split(','))

for i in data:
    for s in i:
        l_data.append(s.replace('\t','').replace('	',',').replace('.      ',',').replace('      ',',').replace('   ',',').replace('  ',',').replace('"','').split(','))

for e in l_data:
    for i,s in enumerate(e[:-1]):
        try:
            e[i] = float(s)
        except:
            print(e)
print(l_data)
```
