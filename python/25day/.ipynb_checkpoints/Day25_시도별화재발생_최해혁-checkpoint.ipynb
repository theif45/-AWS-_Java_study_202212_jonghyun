```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_fname = 'D2Coding-Ver1.3.2-20180524-all.ttc'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)

pd_data = pd.read_csv('시도별화재발생.csv', encoding='cp949')
```

```python
pd_data.loc[pd_data['사망 (명)'] == '-', '사망 (명)'] = '0' 
```

```python
pd_data['사망 (명)'] = pd_data['사망 (명)'].astype(np.int64)
```

```python
pd_data.dtypes
```

```python
filter1 = pd_data['행정구역별'] == '전국'
filter2 = pd_data['행정구역별'] != '전국'
pd_all = pd_data[filter1]
pd_entry = pd_data[filter2]
```

```python
sns.histplot(data= pd_all, x = '행정구역별', y = '이재가구수 (가구)')
```

```python
fig1 = plt.figure()
a = fig1.add_subplot()
plt.xticks(rotation = 270)
sns.lineplot(data= pd_entry, x = '행정구역별', y = '사망 (명)', ax = a)

```

```python
plt.xticks(rotation = 270)
sns.lineplot(data= pd_entry, x = '시점', y = '사망 (명)')
```

```python
sns.jointplot(
    data=pd_entry, x="건수 (건)", y="사망 (명)", xlim=(0, 6000), ylim=(0, 60),
    marker="+", s=100, marginal_kws=dict(bins=25, fill=False),
)
```

```python
sns.jointplot(data=pd_entry, x="건수 (건)", y="사망 (명)", height=5, ratio=2, marginal_ticks=True)
```

```python
g = sns.jointplot(data=pd_entry, x="건수 (건)", y="사망 (명)")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
```

```python
g = sns.JointGrid()
x, y = pd_entry["건수 (건)"], pd_entry["사망 (명)"]
sns.scatterplot(x=x, y=y, ec="b", fc="none", s=100, linewidth=1.5, ax=g.ax_joint)
sns.histplot(x=x, fill=False, linewidth=2, ax=g.ax_marg_x)
sns.kdeplot(y=y, linewidth=2, ax=g.ax_marg_y)
```

```python
g = sns.JointGrid(data=pd_entry, x="건수 (건)", y="부상 (명)", hue="행정구역별")
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data=pd_entry, x="건수 (건)", y="부상 (명)", xlim=(0, 8000), ylim=(0, 400))
g.plot(sns.scatterplot, sns.histplot)
g.refline(x=45, y=16)
```

```python
g = sns.JointGrid(data=pd_entry, x="이재가구수 (가구)", y="이재민수 (명)", xlim=(0, 300), ylim=(0, 600))
g.plot(sns.scatterplot, sns.histplot)
g.refline(x=45, y=16)
```

```python
g = sns.JointGrid(data=pd_entry, x="이재가구수 (가구)", y="이재민수 (명)", xlim=(0, 300), ylim=(0, 600))
g.plot(sns.regplot, sns.boxplot)
```

```python
g = sns.JointGrid(data=pd_entry, x="건수 (건)", y="이재가구수 (가구)", hue="시점")
g.plot(sns.scatterplot, sns.histplot)
```

# 위의 결과를 토대로 알 수 있던 것 들

🔥 가장 많이 알아본 것은 사건이 일어난 건수 당 사망 인원

⭐️ 시점을 주어서 어떠한 년도에 사건이 얼마나 일어났고 그로 인해 사망한 인원이 양의 상관관계를 가지는지 보았다.

⚡️그 결과 양의 상관관계를 가지고 있는 것이 확인이 되었고 다음으로는 건수당 이재 가구수 의 관계
    그리고 이재가구수 당 이재민수의 관계 / 건수 당 부상의 인원 관계 등으로 다양하게 그려보았다.
    
✅거의 같은 결과로 건수에 비례해서 부상/ 이재민수/ 이재가구수등 양의 상관관계를 가지는 것을 확인하였다.


## 좀더 체크 해볼 것

- x축을 건수로 주면서 느꼈던 것은 해당 건수가 보통 4000건 아래에 집합되어 있기에 max를 6000으로 두어 그래프를 그려도 무관 할 거같다.
- 위에 찍어보았지만 재산피해관련은 천원단위로 그래프로 그리게되면 높은 숫자로 집계되기에 단위를 만단위로 바꾸거나 백만단위로 바꿔서 해볼것
- 재산 피해당 부동산과 동산의 피해가 어떻게 다른지 알아보려고 했으나 그래프에서 크게 느껴지는 것이 없었음
