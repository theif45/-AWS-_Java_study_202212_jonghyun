```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

class GraphGUI:
    def __init__(self, master):
        self.master = master
        master.title("Graph GUI")

        # 파일 불러오기 버튼 생성
        self.file_button = tk.Button(master, text="Select File", command=self.load_data)
        self.file_button.pack()

        # 데이터 선택
        self.x_label = tk.Label(master, text="X-Axis")
        self.x_label.pack()
        self.x_entry = tk.Entry(master)
        self.x_entry.pack()

        self.y_label = tk.Label(master, text="Y-Axis")
        self.y_label.pack()
        self.y_entry = tk.Entry(master)
        self.y_entry.pack()

        # 데이터 컬럼 리스트 보이기
        self.column_listbox = tk.Listbox(master)
        self.column_listbox.pack()

        # 그래프 출력 버튼 생성
        self.graph_button = tk.Button(master, text="Show Graph", command=self.show_graph)
        self.graph_button.pack()

        # 그래프 출력 공간 생성
        self.fig, self.ax = plt.subplots()
        self.canvas = FigureCanvasTkAgg(self.fig, master=root)
        self.canvas.get_tk_widget().pack()

    def load_data(self):
        # 파일 선택
        self.filename = filedialog.askopenfilename()

        # 파일 불러오기
        self.df = pd.read_csv(self.filename, encoding='cp949', header=1)

        # 데이터 컬럼 리스트 출력
        self.column_listbox.delete(0, tk.END)
        for column in self.df.columns:
            self.column_listbox.insert(tk.END, column)

    def show_graph(self):
        # x, y 데이터 선택
        x_label = self.x_entry.get()
        y_label = self.y_entry.get()
        x_data = self.df[x_label]
        y_data = self.df[y_label]

        # 그래프 출력
        self.ax.clear()
        self.ax.plot(x_data, y_data)
        self.ax.set_xlabel(x_label)
        self.ax.set_ylabel(y_label)
        self.canvas.draw()

root = tk.Tk()
my_gui = GraphGUI(root)
root.mainloop()
```
