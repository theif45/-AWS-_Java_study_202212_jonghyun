```python
def percentage(data_list,name_list):
    for i,e in enumerate(data_list):
        print(name_list[i],round((e/sum(data_list)*100),2))
        
def get_numbers(data_list, year, month):
    for e in data_list:
        if '{}.{}'.format(year,month) in e:
            return e
    return None
```

```python
import pickle
csv_data=[]
line_no=0
with open('국적별_외국인_현황_20230207151205.csv') as f:
    for line in f:
        line_no = line_no + 1
        if line_no == 2:
            continue
        t = line[:-1].split(',')
        del t[1]
        csv_data.append(t)

for e in csv_data:
    e[0] = e[0].strip('"')

for e in csv_data[1:]:
    for i, value in enumerate(e[1:]):
        e[i+1] = int(value)
        
# print(csv_data)
with open('2020_10-2022_11.pickle', 'wb') as f:
    pickle.dump(csv_data, f)
```

```python
# print(csv_data[:10])

count_list = get_numbers(csv_data, 2022, 11)
count_list = count_list[1:]
country_list = csv_data[0][1:]
print(count_list, len(count_list))
print(country_list, len(country_list))

percent_list=[]
for n in count_list:
    percent_list.append(round(((n*100)/sum(count_list)),2))
print(percent_list)

count_dict=dict(sorted(zip(country_list,count_list), key=lambda x:x[1], reverse=True))
percent_dict=dict(sorted(zip(country_list, percent_list), key=lambda x:x[1], reverse=True))
print(count_dict)
print(percent_dict)
# percentage(count_list,country_list)
```
