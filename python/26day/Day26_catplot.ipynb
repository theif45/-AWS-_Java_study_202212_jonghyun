```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd_tips = sns.load_dataset('tips')
print(pd_tips)
```

```python
sns.catplot(data=pd_tips, x='day', y='total_bill', palette='Set2', jitter=0.2, hue='smoker', dodge=True)
sns.relplot(data=pd_tips, x='day', y='total_bill')
```

```python
sns.catplot(data=pd_tips, x='day', y='total_bill', palette='Set2', hue='smoker', dodge=True , kind = 'swarm')
```

```python
sns.catplot(data=pd_tips, x='total_bill', y='day', palette='Set2', hue='smoker', dodge=True , kind = 'swarm')
```

```python
sns.catplot(data=pd_tips, x='day', y='total_bill', palette='Set2', hue='smoker', dodge=True , kind = 'box')
```

```python
a = sns.catplot(data=pd_tips, x='day', y='total_bill', palette='Set2',dodge=False, kind ='bar', estimator=np.sum, hue='smoker', ci=None)
```

```python
g=sns.catplot(data=pd_tips, x='day', y='total_bill', palette='pastel', hue='smoker', split=True , kind = 'violin', inner='stick')
sns.swarmplot(data=pd_tips, x='day', y='total_bill', ax=g.ax, color='black')
```

```python
species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count
    print(boolean, bottom)

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")
```
