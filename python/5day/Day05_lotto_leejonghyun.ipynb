```python
csv_data=[]
with open('lott.CSV') as f:
    for line in f:
        csv_data.append(line[:-1].split(','))
print(csv_data[:3])
int(csv_data[0][-1])
```

```python
# 1. 전체 회차를 통틀어 빈도수(mode): 숫자별 카운트 (몇번 1-45)
# 2. 회차별 - 평균값, 표준편차
n_data = []
for lotto_data in csv_data:
    for i in lotto_data:
        n_data.append(int(i))
set_data = set(n_data)
k_data = []
v_data = []
for i in set_data:
    k_data.append(i)
    v_data.append(n_data.count(i))
dict1 = dict(zip(k_data,v_data))

m = max(dict1.values())  
for k, v in dict1.items():
    if m == v:
        print(k,v)
    
```

```python
import math

def mean(num_list):
    return sum(num_list)/len(num_list)

def dev(num_list):
    m = mean(num_list)
    return [x-m for x in num_list]

def var(num_list):
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d])/(n-1)

def stdev(num_list):
    return math.sqrt(var(num_list))
```

```python
for e in csv_data:
    for i, s in enumerate(e):
        e[i] = int(s)

for i,e in enumerate(csv_data):
    print('회차:', i+1, '평균:', mean(e), '표준편차:', stdev(e))
```
